```julia
include("../src/CosmoTools.jl")
import .CosmoTools
```

```julia
using Plots

z = 10 .^range(-2, 6, 100) 
p = plot(xscale=:log10, xlabel="z", ylabel="Abundance")
plot!(p, z, Ω.(Radiation, z), label="Radiation")
plot!(p, z, Ω.(Matter, z), label="Matter")
plot!(p, z, Ω.(DarkEnergy, z), label="DarkEnergy")
```

```julia
z = 10 .^range(-2, 6, 100)
p = plot(xscale=:log10, yscale=:log10, xlabel="z", ylabel="Energy density [Msun / Mpc^3]")
plot!(p, z, mean_ρ.(Radiation, z), label="Radiation")
plot!(p, z, mean_ρ.(Matter, z), label="Matter")
plot!(p, z, mean_ρ.(DarkEnergy, z), label="DarkEnergy")
```

```julia
k = 10 .^range(-4, 2, 100)
p = plot(xscale=:log10, yscale=:log10, xlabel="k [Mpc^{-1}]", ylabel="Matter power spectrum [Mpc^3]")
plot!(p, k, matter_power_spectrum.(k), label="ΛCDM")
```

```julia
M = 10 .^range(-10, 20, 100)
p = plot(xscale=:log10, xlabel="M [Msun]", ylabel="Smoothed standard deviation σ(M)")
plot!(p, M, σ_mps_M.(M, SharpKWindow), label="Sharp-K")
plot!(p, M, σ_mps_M.(M, TopHatWindow), label="TopHat")
plot!(p, M, σ_mps_M.(M, GaussianWindow), label="Gaussian")
```
